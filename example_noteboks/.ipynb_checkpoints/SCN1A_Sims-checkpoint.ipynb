```python
import matplotlib.pyplot as plt
import numpy as np
from importlib import reload
from scipy.integrate import odeint

import sodmod as sm
from sodmod import params as pr
from sodmod import cells as cl
```

```python
# adapted from https://gist.github.com/...
#    giuseppebonaccorso/60ce3eb3a829b94abf64ab2b7a56aaef

# Solve ODE system
#---------------------------------------------------------------------------
np.random.seed(1000)
Y0 = [-40,1.,0.,0.]
T  = np.linspace(0,100,10000)
p  = pr.params('TI37')
Vy = odeint(cl.IN, Y0, T, args=(p,))
```

```python
reload(sm.incurr)
reload(sm.cells)

# Solve ODE system
#--------------------------------------------------------------------
np.random.seed(1000)
Y0 = [-40,1.,0.,0.]
T  = np.linspace(0,150,10000)

fig, ax = plt.subplots(2,1, figsize=(12, 14))
Vy      = odeint(cl.IN, Y0, T, args=(pr.params('WT37'),))
ax[0].plot(T, Vy[:,0], 'k')    

Vy      = odeint(cl.IN, Y0, T, args=(pr.params('AS37'),))
ax[0].plot(T, Vy[:,0]-100, 'b')  

Vy      = odeint(cl.IN, Y0, T, args=(pr.params('TI37'),))
ax[0].plot(T, Vy[:,0]-200, 'r') 
```

```python
reload(sm.params)
def F_m(V,V2_m,s_m):return 1 / (1 + np.exp(-(np.divide(V-V2_m,s_m))))
def F_h(V,V2_h,s_h):return 1 / (1 + np.exp((np.divide(V-V2_h,s_h))))

fig, ax = plt.subplots(1,1, figsize=(12, 8))
V = np.array(range(-80,20,1))

conds = ['WT37', 'AS37', 'TI37']
cols  = ['k','b','r']

for ci in range(len(conds)):
    p = pr.params(conds[ci])
    ax.plot(V, F_m(V, p['V2_m'], p['s_m']), cols[ci])
    ax.plot(V, F_h(V, p['V2_h'], p['s_h']), cols[ci])

```

```python
# Input stimulus
from sba_mod_input import Id
Idv = [Id(t) for t in T]
snames = params()['snames_IN']

fig, ax = plt.subplots(figsize=(12, 7))
ax.plot(T, Idv)
ax.set_xlabel('Time (ms)')
ax.set_ylabel(r'Current density (uA/$cm^2$)')
ax.set_title('Stimulus (Current density)')
plt.grid()

# Neuron potential
fig, ax = plt.subplots(figsize=(12, 7))
ax.plot(T, Vy[:, snames.index('Vm')])
ax.set_xlabel('Time (ms)')
ax.set_ylabel('Vm (mV)')
ax.set_title('Neuron potential with two spikes')
plt.grid()

# Trajectories with limit cycles
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(Vy[:, snames.index('Vm')], Vy[:, snames.index('m_K')], label='Vm - n')
ax.plot(Vy[:, snames.index('Vm')], Vy[:, snames.index('m_Na')], label='Vm - m')
ax.set_title('Limit cycles')
ax.legend()
plt.grid()
```
